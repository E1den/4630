```
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
```

## Simple function from R^2 to R and local minima

```
f = lambda x,y: (x**2 +y - 11) ** 2 + (x + y**2 -7)**2

x = np.linspace(-5,5)
y = np.linspace(-5,5)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.show()

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(60)

z = f(x,y)
ax.contour3D(X, Y, Z, 50, cmap='binary')

min_x = [3, -2.805, -3.779, 3.584]
min_y = [2, 3.131, -3.283, -1.848]

ax.scatter(min_x,min_y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

## Simple function from R^2 to R and local maxima

```
f = lambda x,y: -x**2-y**2

x = np.linspace(-5,5)
y = np.linspace(-5,5)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.show()

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(60)

z = f(x,y)
ax.contour3D(X, Y, Z, 50, cmap='binary')

max_x = [0]
max_y = [0]

ax.scatter(max_x,max_y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

## Simple function from R^2 to R and neither a minimum or maximum

```
f = lambda x,y: 0*x+0*y

x = np.linspace(-5,5)
y = np.linspace(-5,5)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.show()

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(60)

z = f(x,y)
ax.contour3D(X, Y, Z, 50, cmap='binary')

neither_x =  [0] #All of the points
neither_y = [0]  #All of the points

ax.scatter(neither_x,neither_y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
